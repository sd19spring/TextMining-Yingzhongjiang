Project Overview: 


For this project I used Gutenberg for my data.
Specifically the ten tragedies by Shakespeare.
The goal for my project is to find common trends in sentiment
by act within the playsself.

Implementation:
After downloading files from Gutenberg, my mini-project focuses
on two primary functions: reshuffling Shakespeare's work into
a format that can be sentiment analyzed by act, and then importing
nltk to analyze the acts.

One design decision is seen in the beginning of the code,
where split is not by the traditional " ". This ironically took
many days (about 3 to 4) of iterating and staring at my computer
wondering if I made a good decision. Original iterations involved
creating a list seperated into acts; I fiddled with creating empty
lists and noodling the txt files through for loops to
ultimately append to said empty lists. However thanks to a strike of
inspiration from unknown source in the middle of coding alone in a dark room,
I was able to find a simpler version, and decided to just do
str.split("ACT"). It was blissfully shorter, and thus my choice on
adopting it for my project.

Results:
This Reflection will be sent along with other files. One such file is a
plotted graph of the sentiment over acts. Within the graph, it was
particularly interesting to realize that Shakespearian tragedies are,
at least to a modern computer, not tragic. Furthermore, over each of the
five acts within each play, the compound sentiment is very consistent and
hovers around a 1.0. The attached image of the plotted graph had its y-axis
values were adjusted to 1.0 and 0.999 to show slight variances among plays.

The other interesting find is in addition to each act within a play having
minimal variance in sentiment, thus creating a straight line parallel to the
x-axis, each play has a very minimal overall sentiment variance when compared
to other plays. The attached image of the plotted graph had its y-axis
values were adjusted to 1.0 and 0.999 to show slight variances among plays.



Alignment:

There is a fair amount of alignment between what I set out to explore, the data,
and the tools used to carry out the project. Although it was dissappointing todo
see minimal variance over act of a Shakespearian tragedy, it was very interesting
to find that Shakespeare was able to maintain an overall consistent level
of sentiment throughout each play.

This project came to be from talking with Ashley and a few other classmates on
tentative ideas for potential project work. Originally it was to find if
sentiment of Disney movies changed over time, especially with its acquisition
of Pixar. Thanks to Melissa, I was saved from diving too deep into a murky place of
legal issues pertaining to copyright, and chose to redirect myself to the word

I imagined my data would reveal the key to writing great, timeless tragic plays through
the use of careful sentiment alterations by act within a play. And this would be one
of the many leads into recreating timeless classics. Thankfully Shakespeare wrote
a limited number of tragedies, so data collection was not as bad. However, because
each edition of his plays are always tweaked for our understanding, there could be

the coder herself. Due to time limitations, Anne decided to simply focus on plays by
act and not further break Shakespeare's works into each scene; this
could potentially distort our results, and accordingly, potentially give room for error
in Anne's confidence in her discovery.


Reflection:

What could I improve: break plays down into scenes and then use sentiment analysis, incorporate
